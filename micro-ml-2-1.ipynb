[糖尿病数据集](https://scikit-learn.org/stable/datasets/toy_dataset.html#diabetes-dataset)

```python
# The diabetes dataset糖尿病数据集
# 
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
print(X[1])
```

接下来，通过选择数据集的第3列，选择要绘图的该数据集的一部分。您可以使用:操作符选择所有行，然后使用索引（2）选择第3列。您还可以使用reshape(n_rows, n_columns)将数据整形为二维阵列（根据打印的需要）。如果其中一个参数为-1，则自动计算相应的尺寸。

```python
X = X[:, 2]
X = X.reshape((-1,1))
print(X)
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```
